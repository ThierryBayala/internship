```python
from time import sleep

import requests
from bs4 import BeautifulSoup
from kafka import KafkaProducer
import pymongo
from pymongo import MongoClient
import pprint

def publish_message(producer_instance, topic_name, key, value):
    try:
        key_bytes = bytes(key, encoding='utf-8')
        value_bytes = bytes(value, encoding='utf-8')
        producer_instance.send(topic_name, key=key_bytes, value=value_bytes)
        producer_instance.flush()
        print('Message published successfully.')
    except Exception as ex:
        print('Exception in publishing message')
        print(str(ex))


def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['localhost:9092'], api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka')
        print(str(ex))
    finally:
        return _producer

def fect_tweet(dict_object):
    print('Processing..{}'.format(dict_object))
    print()
    try:
        if 'text' in dict_object:
                sleep(3)
                tweet = dict_object['text']
    except Exception as ex:
        print('Exception while accessing raw tweet')
        print(str(ex))
    finally:
        return tweet.strip()


    
def fetch_date(dict_object):
    #print('Processing..{}'.format(dict_object))
    #print()
    try:
        if 'text' in dict_object:
                sleep(1)
                tweet = dict_object['date']
    except Exception as ex:
        print('Exception while accessing raw tweet')
        print(str(ex))
    finally:
        return tweet.strip()

    
    

def get_tweets():
    print("===> Qerying data from MongoDb ...")
    print()
    tweets = []
    client = MongoClient('localhost', 27017)
    db = client.twitter_db
    collection = db.data
    #skip( 20 )
    result = collection.find().skip(40000).limit(4000)
    try:    
        for res in result:
            tweet=fect_tweet(res)
            tweets.append(tweet)
            
    except Exception as ex:
        print('Exception in get_tweet')
        print(str(ex))
    finally:
        return tweets
    
    
def get_dates():
    dates = []
    client = MongoClient('localhost', 27017)
    db = client.twitter_db
    collection = db.data
    result = collection.find().limit(10)
    try:    
        for res in result:
            tweet=fetch_date(res)
            dates.append(tweet)
            
    except Exception as ex:
        print('Exception in get_tweet')
        print(str(ex))
    finally:
        return dates

if __name__ == '__main__':
    all_tweets = get_tweets()
    
    if len(all_tweets) > 0:
        kafka_producer = connect_kafka_producer()
        for tweet in all_tweets:
            publish_message(kafka_producer, 'raw_tweets', 'raw_date', tweet.strip())
        if kafka_producer is not None:
            kafka_producer.close()
```
