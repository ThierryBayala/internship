```python
import json
from time import sleep
from cleaner import *
from bs4 import BeautifulSoup
from kafka import KafkaConsumer, KafkaProducer


def publish_message(producer_instance, topic_name, key, value):
    try:
        key_bytes = bytes(key, encoding='utf-8')
        value_bytes = bytes(value, encoding='utf-8')
        
        producer_instance.send(topic_name, key=key_bytes.lower(), value=value_bytes.lower())
        producer_instance.flush()
        print('Tweet published successfully.')
    except Exception as ex:
        print('Exception in publishing tweet')
        print(str(ex))


def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['localhost:9092'], api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka')
        print(str(ex))
    finally:
        return _producer


def parse(tweet):
    title = str(tweet)
    rec = {}
    try:
        rec = {'tweet': title}

    except Exception as ex:
        print('Exception while parsing')
        print(str(ex))
    finally:
        return json.dumps(rec)

def removePartPunc(doc):
    data = []
    for sent in doc.sents:
        for w in sent:
            tmp=[]
            if 'PUNCT' in w.pos_:
                pass
            elif 'PART' in  w.pos_:
                pass
            elif 'NUM' in w.pos_:
                pass
            
            elif (w.text).strip()=='rt':
                pass
            else:
                tmp.append(w.text)
                tmp.append(w.pos_)
                data.append(tmp)
    tokens = pd.DataFrame(data=data, columns = ['TOKENS','POS'])
    return tokens['TOKENS'].tolist()
    
    

    
if __name__ == '__main__':
    print('===> Cleanning Data ...')
    print()
    cleaned_records = []
    topic_name = 'parsed_tweet'    
    parsed_topic_name = 'cleaned_tweet'

    consumer = KafkaConsumer(topic_name, auto_offset_reset='earliest',
                             bootstrap_servers=['localhost:9092'], api_version=(0, 10), consumer_timeout_ms=1000)
    for msg in consumer:
        record=json.loads(msg.value)
        text=record['tweet']
        doc = nlp(text)
        doc2 = negation_tag(doc)
        if len(doc2)>1:
            out_put=' '.join(doc2)
        else:
            out_put=''.join(doc2)
        
        doc3=nlp(out_put)
        
        out_put=removePartPunc(doc3)
        out_put=' '.join(out_put)
        print(out_put.strip())
        
        result = parse(out_put.strip())
        
        cleaned_records.append(result)

    consumer.close()
    sleep(5)

    if len(cleaned_records) > 0:
        print()
        print('Publishing records...')
        print()
        producer = connect_kafka_producer()
        for rec in cleaned_records:
            
            publish_message(producer, parsed_topic_name, 'cleaned', rec)
```
