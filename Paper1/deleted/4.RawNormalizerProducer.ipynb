```python
from time import sleep
from normalizer import *
from bs4 import BeautifulSoup
from kafka import KafkaConsumer, KafkaProducer


def publish_message(producer_instance, topic_name, key, value):
    try:
        key_bytes = bytes(key, encoding='utf-8')
        value_bytes = bytes(value, encoding='utf-8')
        
        producer_instance.send(topic_name, key=key_bytes.lower(), value=value_bytes.lower())
        producer_instance.flush()
        print('Tweet published successfully.')
    except Exception as ex:
        print('Exception in publishing tweet')
        print(str(ex))


def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['localhost:9092'], api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka')
        print(str(ex))
    finally:
        return _producer


def parse(tweet):
    title = str(tweet)
    rec = {}
    try:
        rec = {'tweet': title}

    except Exception as ex:
        print('Exception while parsing')
        print(str(ex))
    finally:
        return json.dumps(rec)

def lammaProd(doc):
    data = []
    for sent in doc.sents:
        for w in sent:
            tmp=[]
            if 'NOUN' in w.pos_:
                tmp.append(w.lemma_)
                tmp.append(w.pos_)
                data.append(tmp)
                
            else:
                tmp.append(w.text)
                tmp.append(w.pos_)
                data.append(tmp)
    tokens = pd.DataFrame(data=data, columns = ['TOKENS','POS'])
    return tokens['TOKENS'].tolist()
    
    

    
if __name__ == '__main__':
    print('===> Normalizing Data ...')
    print()
    reords_normalized=[]
    normalized_records = []
    
    topic_name = 'cleaned_tweet'    
    parsed_topic_name = 'normalized_tweet'

    consumer = KafkaConsumer(topic_name, auto_offset_reset='earliest',
                             bootstrap_servers=['localhost:9092'], api_version=(0, 10), consumer_timeout_ms=1000)
    for msg in consumer:
        record=json.loads(msg.value)
        text=record['tweet']
        doc = nlp(text)
        out_put=lammaProd(doc)
        out_put=' '.join(out_put)
        
        #print(out_put.strip())
        
        result = parse(out_put)
        normalized_records.append(result)

    consumer.close()
    sleep(5)

    dList=duplication_removal(normalized_records,normalized_records)
    
    for i,text in enumerate(normalized_records):
        if(i not in dList):
            record=json.loads(text)
            text=record['tweet']
            print(text.strip())
            reords_normalized.append(parse(text.strip()))

    if len(reords_normalized) > 0:
        print()
        print('Publishing records...')
        print()
        producer = connect_kafka_producer()
        for rec in reords_normalized:
            publish_message(producer, parsed_topic_name, 'normalized', rec)
```
